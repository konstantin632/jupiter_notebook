```python
t=np.linspace(0,10,51)
t
```

```python
f=np.cos(t)
f
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "Calibri",
}
label_font = {
    "fontsize": 13,
    "family": "Calibri",
}
```

```python
DATASET_PATH = './Downloads/creditcard.csv'
```

```python
df = pd.read_csv(DATASET_PATH, sep=',')
df.head()
```

```python
class_count=df['Class'].value_counts()
class_count
```

```python
class_count.plot(kind="bar")
plt.xlabel('Class',fontdict=label_font)
plt.ylabel('Value counts',fontdict=label_font)
plt.title('Bar chart of Class',fontdict=title_font)


plt.show()
```

```python
class_count.plot(kind="bar", logy=True)

plt.xlabel('Class',fontdict=label_font)
plt.ylabel('Value counts',fontdict=label_font)
plt.title('Bar chart of Class',fontdict=title_font)

plt.show()
```

```python
x1=df.loc[df['Class']==1,'V1']
x2=df.loc[df['Class']==0,'V1']
x1.value_counts()
```

```python
colors = ['red', 'grey']
names = ['Class 1', 'Class 0']
    
plt.hist([x1, x2], bins = 20,density=True,
         color = colors, label=names, alpha=0.5)

plt.legend()
plt.xlabel('V1',fontdict=label_font)
plt.ylabel('distribution density',fontdict=label_font)
plt.title('Histogramme distribution of V1',fontdict=title_font)
```
